```python
# Group by the player who received fouls
fouls_received = fdf.groupby("player")["player_foul_commitedJ"].apply(list).reset_index()

# Count the number of fouls received
fouls_received["foul_count"] = fouls_received["player_foul_commitedJ"].apply(len)

# Find the player committing the most fouls
fouls_received["top_committer"] = fouls_received["player_foul_commitedJ"].apply(lambda x: pd.Series(x).mode()[0])

# Add count of True values in foul_committed_advantage_foul_commitedJ
fouls_received["advantage_count"] = fdf.groupby("player")["foul_committed_advantage_foul_commitedJ"].sum().values

# Add counts of specific cards in foul_committed_card_foul_commitedJ
cards = fdf.groupby("player")["foul_committed_card_foul_commitedJ"].apply(list).reset_index()
cards["yellow_card_count"] = cards["foul_committed_card_foul_commitedJ"].apply(lambda x: x.count("Yellow Card"))
cards["second_yellow_count"] = cards["foul_committed_card_foul_commitedJ"].apply(lambda x: x.count("Second Yellow"))
cards["red_card_count"] = cards["foul_committed_card_foul_commitedJ"].apply(lambda x: x.count("Red Card"))

# Merge the card counts back into the main DataFrame
fouls_received = fouls_received.merge(cards[["player", "yellow_card_count", "second_yellow_count", "red_card_count"]], on="player")

# Sort by foul_count in descending order
fouls_received = fouls_received.sort_values(by="foul_count", ascending=False)

# Display the top 10 players
fouls_received.head(10)
```

```python
sns.histplot(data=fouls_received, x="foul_count")
```

```python
# Group by the player who committed fouls and count occurrences
fouls_commited = fdf["player_foul_commitedJ"].value_counts().reset_index()

# Rename columns for better understanding
fouls_commited.columns = ["player", "foul_count"]

# Sort by foul_count in descending order (if not already sorted)
fouls_commited = fouls_commited.sort_values(by="foul_count", ascending=False)

# Display the top 10 players
fouls_commited.head(10)
```

```python
sns.histplot(data=fouls_commited, x="foul_count")
```
